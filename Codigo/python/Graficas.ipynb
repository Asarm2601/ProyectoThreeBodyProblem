Método de Runge Kutta

```python
import pandas as pd
import plotly
from matplotlib import pyplot as plt
import numpy as np
print(plotly.__version__)
data = pd.read_csv("../cpp/rungekuta.csv")
data2 = pd.read_csv("../cpp/sistemalunatierra.csv")
data3 = pd.read_csv("../cpp/sistemasateliteluna.csv")
datos = pd.read_csv("../cpp/cantidadesconservadas.csv")
```

```python

import plotly.graph_objects as go
import matplotlib
fig = go.Figure()
AU = 1.496e12
fig.add_trace(go.Scatter3d(
    x=data["x1"]/AU,
    y=data["y1"]/AU,
    z=data["z1"]/AU,
    mode='lines',
    name='Cuerpo 1'
))

fig.add_trace(go.Scatter3d(
    x=data["x2"]/AU,
    y=data["y2"]/AU,
    z=data["z2"]/AU,
    mode='lines',
    name='Cuerpo 2'
))

fig.add_trace(go.Scatter3d(
    x=data["x3"]/AU,
    y=data["y3"]/AU,
    z=data["z3"]/AU,
    mode='lines',
    name='Cuerpo 3'
))

fig.update_layout(
    scene=dict(
        xaxis_title="X",
        yaxis_title="Y",
        zaxis_title="Z"
    ),
    width=900,
    height=700
)
fig.show()
AU2 = 1.496e11
fig = go.Figure()
fig.add_trace(go.Scatter3d(
    x=data2["x1"]/AU2,
    y=data2["y1"]/AU2,
    z=data2["z1"]/AU2,
    mode='lines',
    name='Cuerpo 3 respecto a 1'
))
fig.update_layout(
    scene=dict(
        xaxis_title="X",
        yaxis_title="Y",
        zaxis_title="Z"
    ),
    width=900,
    height=700
)
fig.show()
fig = go.Figure()
fig.add_trace(go.Scatter3d(
    x=data3["x1"]/AU2,
    y=data3["y1"]/AU2,
    z=data3["z1"]/AU2,
    mode='lines',
    name='Cuerpo 3 respecto a 2'
))
fig.update_layout(
    scene=dict(
        xaxis_title="X",
        yaxis_title="Y",
        zaxis_title="Z"
    ),
    width=900,
    height=700
)
fig.show()
```

```python
# Tiempo
t = datos["t"]

# Magnitud del momento lineal
P = np.sqrt(datos["p1"]**2 +
            datos["p2"]**2 +
            datos["p3"]**2)

# Magnitud del momento angular
L = np.sqrt(datos["L1"]**2 +
            datos["L2"]**2 +
            datos["L3"]**2)

# ===============================
# Energía
# ===============================
plt.figure(figsize=(8,4))
plt.plot(t, datos["E"], lw=1.5)
plt.xlabel("Tiempo")
plt.ylabel("Energía")
plt.title("Conservación de la Energía")
plt.grid(True)
plt.tight_layout()
plt.show()

# ===============================
# Momento lineal
# ===============================
plt.figure(figsize=(8,4))
plt.plot(t, P, lw=1.5)
plt.xlabel("Tiempo")
plt.ylabel(r"$|\mathbf{P}|$")
plt.title("Magnitud del Momento Lineal")
plt.grid(True)
plt.tight_layout()
plt.show()

# ===============================
# Momento angular
# ===============================
plt.figure(figsize=(8,4))
plt.plot(t, L, lw=1.5)
plt.xlabel("Tiempo")
plt.ylabel(r"$|\mathbf{L}|$")
plt.title("Magnitud del Momento Angular")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Por componentes

# Momento lineal
plt.figure(figsize=(8,4))
plt.plot(t, datos["p1"], label="Px")
plt.xlabel("Tiempo")
plt.ylabel("Momento lineal")
plt.legend()
plt.figure(figsize=(8,4))
plt.plot(t, datos["p2"], label="Py")
plt.plot(t, datos["p3"], label="Pz")
plt.xlabel("Tiempo")
plt.ylabel("Momento lineal")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
# Momento angular
plt.figure(figsize=(8,4))
plt.plot(t, datos["L1"], label="Lx")
plt.plot(t, datos["L2"], label="Ly")
plt.plot(t, datos["L3"], label="Lz")
plt.xlabel("Tiempo")
plt.ylabel("Momento angular")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
